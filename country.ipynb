#### Comparison of Country Area

```python
#1. what are the top 5 biggest country in Area?

fig, ax = plt.subplots(figsize = (16, 5))

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], round(y[i]/10000000,4), ha = 'center')
country = selected_country['Country']
values = selected_country['Area(sq km)']
plt.grid(True)
plt.bar(country , values, color = "brown")
plt.xticks(fontsize = 12 ,rotation = 90)
plt.yticks(fontsize = 12)

plt.xlabel('Countrys', fontsize= 15)

addlabels(country,values)
ax.set_title('Country Area in 10 Million Square Kilometer Units.')
plt.show()

# top 5 biggest country in Area Russia, Canada, United States, China, India
```

#### Current account balance
<p>When a country's current account balance is positive, the country is a net lender to the rest of the world.

 When a country's current account balance is negative, the country is a net borrower from the rest of the world.</p>

```python
#4. countrys which have the account balance in negative?

fig, ax = plt.subplots(figsize = (16, 5))

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], round(y[i]/100000000,2), ha = 'center')
country = selected_country['Country']
values = selected_country['Current account balance']
plt.grid(True)
plt.bar(country , values ,color = ['red' if val < 0 else 'green' for val in values ])
plt.xticks(fontsize = 12 ,rotation = 90)
plt.yticks(fontsize = 12)

plt.xlabel('Countrys', fontsize= 15)

addlabels(country,values)
ax.set_title('Country Current account balance in 10 Million Units.')
plt.show()

#United States, united kingdom, mexico, turkey,France, italy  have the account balance in negative
```

#### Export vs Import


```python
#6. Are there more countries that export more than they import?

export =[]
imports = []

for index, countryy in selected_country.iterrows():
    if countryy['Exports'] > countryy['Imports']:
        export.append(countryy['Country'])
    elif countryy['Imports'] > countryy['Exports']:
        imports.append(countryy['Country'])

print(f'Here are the countries with higher exports ,{export},{len(export)}')
print(f'here are the countrse with higher  imports{imports},{len(imports)}')


#Yes, there are more countries which export more than import.
# counties with higher exports are 10
# & countries with higher imports are 9
```

### Finding the best country to live in

```python
selected_country.head()
```

```python
forPEople = selected_country[['Country','GDP - per capita', 'Inflation rate (consumer prices)(%)','Infant mortality rate(deaths/1000 live births)','Life expectancy at birth(years)','Total fertility rate(children born/woman)','Unemployment rate(%)']]

forPEople 
```

#### GDP - per capita

```python
forPEople = forPEople.sort_values('GDP - per capita',ascending=False).reset_index(drop=True)
forPEople['Score-GDPPC']= forPEople.index
forPEople
```

#### Inflation rate (consumer prices)(%)

```python
forPEople = forPEople.sort_values('Inflation rate (consumer prices)(%)',ascending=True).reset_index(drop=True)
forPEople['Score-Inflation']= forPEople.index
forPEople
```

#### Infant mortality rate(deaths/1000 live births)

```python
forPEople = forPEople.sort_values('Infant mortality rate(deaths/1000 live births)',ascending=True).reset_index(drop=True)
forPEople['Score-infant-mortality']= forPEople.index
forPEople
```

#### Life expectancy at birth(years)

```python
forPEople = forPEople.sort_values('Life expectancy at birth(years)',ascending=False).reset_index(drop=True)
forPEople['Score-life-expectency']= forPEople.index
forPEople
```

#### Total fertility rate(children born/woman)

```python
forPEople = forPEople.sort_values('Total fertility rate(children born/woman)',ascending=False).reset_index(drop=True)
forPEople['Score-Total-fertility']= forPEople.index
forPEople
```

#### Unemployment rate(%)

```python
#country with lowest unemploymwnt rate?
forPEople = forPEople.sort_values('Unemployment rate(%)',ascending=True).reset_index(drop=True)
forPEople['Score-Unemployment']= forPEople.index
forPEople

#The United Arab Emirates has the lowest unemployment rate.
```

```python
forPEople['Ranking'] = forPEople['Score-GDPPC'] + forPEople['Score-Inflation'] +forPEople['Score-infant-mortality'] +forPEople['Score-life-expectency'] +forPEople['Score-Total-fertility']+forPEople['Score-Unemployment']
forPEople.sort_values('Ranking', ascending=True).reset_index()[['Country']]
```
